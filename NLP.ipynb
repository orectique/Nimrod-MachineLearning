```python
import pandas as pd 
import numpy as np
import spacy
```

```python
en = spacy.load('en_core_web_sm')
sw_spacy = en.Defaults.stop_words
print(sw_spacy)
```

```python
df = pd.read_csv('title_category.csv')
df.head()
```

```python
cleaned_column = []
for row in df['Content']:
    temp = ''
    for word in str(row).split():
        if word not in sw_spacy:
            temp += ' ' + word
    cleaned_column.append(temp)


df['Cleaned'] = cleaned_column
```
