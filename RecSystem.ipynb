```python
import pandas as pd
import random
```

```python
products = pd.read_csv('cleaned.csv')
```

```python
prods = set(products.ProductId)
people = set(products.UserId)
```

### Bunch people's comments and product comments into separate dataframes

```python
temp1 = []
temp2 = []
temp3 = []

for product in prods:
    temp = products[products['ProductId'] == product]
    temp1.append(product)
    temp2.append(temp['Content'].str.cat(sep = ' '))
    temp3.append(temp.Title[0])

df_prods = pd.DataFrame({'ProductId': temp1, 'Content': temp2, 'Title': temp3})

temp1 = []
temp2 = []

for user in people:
    temp = products[products['UserId'] == user]
    temp1.append(user)
    temp2.append(temp['Content'].str.cat(sep = ' '))

df_users = pd.DataFrame({'UserId': temp1, 'Content': temp2})
    
```
